```python
scene = pywavefront.Wavefront(
  '../data/obj/1a0gA02.obj',
  strict=True,
  encoding="iso-8859-1",
  parse=True,
  collect_faces=True
)
```

```python
print("Vertices:")
print(len(scene.vertices))

print("Meshes:")
print(len(scene.mesh_list))

mesh = scene.mesh_list[0]
print("Faces:")
print(len(mesh.faces))

mesh.faces
```

```python
points = np.array(scene.vertices)
faces = np.array(mesh.faces)
order = 20
scale_input = True
decimate_fraction = 0
decimate_smooth = 0
verbose = False
descriptors = zernike_moments(points, faces, order, scale_input, decimate_fraction, decimate_smooth, verbose)
```

```python
print(len(descriptors))
px.bar(descriptors)
```

## Experiment with Mesh Decimation

```python
# load high-resolution mesh:
m = mr.loadMesh(mr.Path("../data/obj/16pkA02.obj"))
mr.saveMesh(m, mr.Path("./highres.stl"))

# # decimate it with max possible deviation 0.5:
settings = mr.DecimateSettings()
settings.maxError = 0.2
result = mr.decimateMesh(m, settings)
print(result.facesDeleted)
# # 708298
print(result.vertsDeleted)
# # 354149

mr.saveMesh(m, mr.Path("./lowres.stl"))
```
